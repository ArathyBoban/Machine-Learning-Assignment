```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
print(f"Dataset shape: {X.shape}")
print(f"First five rows:\n{X.head()}")
```

```python
# 2.Clustering Algorithm Implementation
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
X['KMeans_Cluster'] = kmeans.fit_predict(X)
plt.figure(figsize=(8, 6))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=X['KMeans_Cluster'], cmap='viridis', edgecolor='k', s=100)
plt.title("KMeans Clustering on Iris Dataset")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.colorbar(label="Cluster")
plt.show()
        
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
linked = linkage(X.iloc[:, :-1], method='ward')
plt.figure(figsize=(10, 7))
dendrogram(linked, truncate_mode='lastp', p=10, show_leaf_counts=True)
plt.title("Dendrogram for Hierarchial Clustering")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
agg_cluster = AgglomerativeClustering(n_clusters=3, linkage='ward')
X['Hierarchical_Cluster'] = agg_cluster.fit_predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=X['Hierarchical_Cluster'], cmap='rainbow',edgecolor='k', s=100)
plt.title("Hierarchical Clustering on Iris Dataset")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.colorbar(label="Cluster")
plt.show()
```
